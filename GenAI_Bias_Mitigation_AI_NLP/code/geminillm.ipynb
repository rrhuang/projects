```python
import time
```

```python
responses = []
for i in range(len(df)):
    print(i)
    prompt = df['Query'][i]
    response = model.generate_content(prompt)
    responses.append(response.text)
    time.sleep(1.5)
```

```python
len(responses)
```

```python
df['Response-Gemini'] = responses
```

```python
new_responses = []
for i in range(len(df)):
    print(i)
    prompt = df['New Query'][i]
    new_response = model.generate_content(prompt)
    new_responses.append(new_response)
    time.sleep(1)
```

```python
new_responses
```

```python
new_responses[0].text
```

```python
def parse_text(resp):
    try:
        ret = resp.text
        return ret
    except:
        return "Error from Gemini"
    
text_of_new_responses = [parse_text(text) for text in new_responses]
```

```python
df['New Response-Gemini'] = text_of_new_responses
```

```python
df = df.drop(columns= ['Unnamed: 0.2', 'Unnamed: 0.1','level_0','index','Unnamed: 0'])
```
