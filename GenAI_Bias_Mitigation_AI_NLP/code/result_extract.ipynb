```python
def extract_last_word_of_first_sentence(sentence):
    first_sentence = sentence.split('.')[0] + '.'
    match = re.search(r'\b(\w+)\b\'?\.$', first_sentence)
    if match:
        return match.group(1)
    else:
        return None
```

```python
df['LLAMA-OGQuery-Answer'] = df['Response-LLAMA'].apply(lambda x: extract_last_word_of_first_sentence(x))
```

```python
df['GPT-OGQuery-Answer'] = df['Response-GPT'].apply(lambda x: extract_last_word_of_first_sentence(x))
df['Gemini-OGQuery-Answer'] = df['Response-Gemini'].apply(lambda x: extract_last_word_of_first_sentence(x))
```

```python
df['GPT-OG-Correct'] = df['GPT-OGQuery-Answer'] == df['Prof']
df['LLAMA-OG-Correct'] = df['LLAMA-OGQuery-Answer'] == df['Prof']
df['Gemini-OG-Correct'] = df['Gemini-OGQuery-Answer'] == df['Prof']
```

```python
def extract_last_word_of_last_sentence(sentence):
    sentences = [s.strip() for s in sentence.split('.') if s.strip()]
    if sentences:
        last_sentence = sentences[-1]
        match = re.search(r'\b(\w+)\b\'?$', last_sentence)
        if match:
            return match.group(1).lower()
    return None
```

```python
df['LLAMA-NewQuery-Answer'] = df['New Response-LLAMA'].apply(lambda x: extract_last_word_of_last_sentence(x))
df['GPT-NewQuery-Answer'] = df['New Response-GPT'].apply(lambda x: extract_last_word_of_last_sentence(x))
df['Gemini-NewQuery-Answer'] = df['New Response-Gemini'].apply(lambda x: extract_last_word_of_last_sentence(x))
```

```python
df['GPT-New-Correct'] = df['GPT-NewQuery-Answer'] == df['Prof']
df['LLAMA-New-Correct'] = df['LLAMA-NewQuery-Answer'] == df['Prof']
df['Gemini-New-Correct'] = df['Gemini-NewQuery-Answer'] == df['Prof']
```
